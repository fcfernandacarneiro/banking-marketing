```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = 'C:/Users/clovi/AppData/Local/Packages/PythonSoftwareFoundation.Python.3.11_qbz5n2kfra8p0/LocalCache/local-packages/Python311/Scripts/bank-additional-full.csv' 

data = pd.read_csv(file_path, delimiter=';')
```

```python
print(data.describe())
```

# Exploration

```python
#Calculating the avarage adoption rate
total_rows = len(data)
count_yes = data.query('y == "yes"').shape[0]

# Make sure to handle the case where count_yes is 0 to avoid ZeroDivisionError
if count_yes > 0:
    ratio = count_yes / total_rows 
    print(f"Avg adoption rate: {ratio:.2f}")
else:
    print("No rows where y is 'yes'")
```

```python
# Display unique values in 'job' column
print(data['job'].unique())
```

```python
grouped_counts = data.groupby('job').size()
print(grouped_counts)
```

```python
# Group and sort the data
grouped_counts = data.groupby('marital').size().sort_values(ascending=False)

# Creating a bar chart
plt.figure(figsize=[10,6])
bars = plt.bar(grouped_counts.index, grouped_counts.values, color='skyblue')

# Adding title and labels
plt.title('Count of rows per Job Category')
plt.xlabel('Job Category')
plt.ylabel('Count')
plt.xticks(rotation=45)  # Rotating the x-axis labels for better readability

# Adding data labels on top of the bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.05, round(yval, 2), va='center', ha='center', fontsize=8)

# Displaying the plot
plt.tight_layout()  # Adjusts subplot params so that the subplot fits into the figure area
plt.show()
```
